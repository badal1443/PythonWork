```python
#Working with Pandas library
import pandas as pd
df1= pd.DataFrame(columns=["Column1","Column2"])
df1.columns=["First","Second"]
df1
```

```python
d={"First":[1,2,3],"Second":[6,7,8]}
df1=pd.DataFrame(data=d)
df1
```

```python
df2=pd.DataFrame({"First":["a","b","c"],"Second":["x","y","z"]})
df2
```

```python
##Row-wise concat
#axis=0 for rowwise concat
df3=pd.concat([df2,df1],axis=0)
df3
```

```python
##Re-setting the index to be in sequence for all rows
df3.reset_index(drop=True,inplace=True)
df3
```

```python
## Create a dataframe with 1 column only
df4= pd.DataFrame({"Third":["Hello","Hi","Jola"]})
df4
```

```python
## Concat df3 with df4 columnwise, meaning add df4 as another column to existing df3 dataframe
df5=pd.concat([df3,df4],axis=1)
df5
```
